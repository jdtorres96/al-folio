Point 1: Generate 100 and 50,000 normal random variables with $\mu$ = 2 and $\sigma$ = 1. Plot the histograms of these random number and draw normal density functions with the given parameters on top of them.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

# Define parameters
mean = 2
std_dev = 1
```

100 random normal variables

```python
rnv_100 = np.random.normal(mean, std_dev, 100)

plt.hist(rnv_100, density=True, label='Histogram')
x_axis = np.linspace(mean - 2.96*std_dev, mean + 2.96*std_dev, 100)
y_axis = norm.pdf(x_axis, mean, std_dev)
plt.plot(x_axis, y_axis, label='Density Function')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('n=100')
plt.legend()

plt.savefig('rnv100.pdf')
plt.show()
```

50,000 normal random variables

```python
rnv_50000 = np.random.normal(mean, std_dev, 50000)

plt.hist(rnv_50000, bins=100, density=True, label='Histogram')
x_axis = np.linspace(mean - 2.96*std_dev, mean + 2.96*std_dev, 100)
y_axis = norm.pdf(x_axis, mean, std_dev)
plt.plot(x_axis, y_axis, label='Density Function')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('n=50,000')
plt.legend()

plt.savefig('rnv50000.pdf')
plt.show()
```
